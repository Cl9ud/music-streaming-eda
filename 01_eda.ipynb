```python
# =========================
# 01_eda.ipynb - Exploración de datos
# =========================

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Configuración de gráficos
sns.set(style="whitegrid")
plt.rcParams["figure.figsize"] = (10,6)

# -------------------------
# 1. Cargar dataset
# -------------------------
df = pd.read_csv("data/Global_Music_Streaming_Listener_Preferences.csv")

# -------------------------
# 2. Limpieza básica
# -------------------------
# Normalizar nombres de columnas
df.columns = df.columns.str.lower().str.replace(" ", "_")

# Eliminar duplicados
df = df.drop_duplicates()

# Revisar valores faltantes
print("Valores faltantes por columna:\n", df.isnull().sum())

# -------------------------
# 3. Información general
# -------------------------
print("\nPrimeras filas:\n", df.head())
print("\nInfo del dataset:\n")
df.info()
print("\nEstadísticas descriptivas:\n", df.describe(include='all'))

# -------------------------
# 4. Visualizaciones iniciales
# -------------------------
# Top 10 artistas más escuchados
if "artist_name" in df.columns:
    top_artists = df["artist_name"].value_counts().head(10)
    sns.barplot(x=top_artists.values, y=top_artists.index, palette="viridis")
    plt.title("Top 10 artistas más escuchados")
    plt.xlabel("Cantidad de registros")
    plt.ylabel("Artista")
    plt.show()

# Top 10 géneros más populares
if "genre" in df.columns:
    top_genres = df["genre"].value_counts().head(10)
    sns.barplot(x=top_genres.values, y=top_genres.index, palette="magma")
    plt.title("Top 10 géneros más populares")
    plt.xlabel("Cantidad de registros")
    plt.ylabel("Género")
    plt.show()

# Top 10 países con más registros
if "country" in df.columns:
    top_countries = df["country"].value_counts().head(10)
    sns.barplot(x=top_countries.values, y=top_countries.index, palette="cubehelix")
    plt.title("Top 10 países con más registros")
    plt.xlabel("Cantidad de registros")
    plt.ylabel("País")
    plt.show()
```
