```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import os
os.chdir("/Users/juancastroandrade23/music-streaming-project")

sns.set(style="whitegrid")
plt.rcParams["figure.figsize"] = (10,6)
```

```python
df = pd.read_csv("data/Global_Music_Streaming_Listener_Preferences.csv")
df.head()
```

```python
import os
os.getcwd()
```

```python
# Normalizar nombres de columnas
df.columns = df.columns.str.lower().str.replace(" ", "_")

# Eliminar duplicados
df = df.drop_duplicates()

# Revisar valores faltantes
print("Valores faltantes por columna:\n", df.isnull().sum())
```

```python
# Información general
df.info()

# Estadísticas descriptivas
df.describe(include='all')
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

top_artists = df["most_played_artist"].value_counts().head(10)
sns.barplot(x=top_artists.values, y=top_artists.index, palette="viridis")
plt.title("Top 10 artistas más escuchados")
plt.xlabel("Cantidad de registros")
plt.ylabel("Artista")
plt.show()
```

```python
df.columns
```

```python
top_genres = df["top_genre"].value_counts().head(10)
sns.barplot(x=top_genres.values, y=top_genres.index, palette="magma")
plt.title("Top 10 géneros más populares")
plt.xlabel("Cantidad de registros")
plt.ylabel("Género")
plt.show()

```

```python
top_countries = df["country"].value_counts().head(10)
sns.barplot(x=top_countries.values, y=top_countries.index, palette="cubehelix")
plt.title("Top 10 países con más registros")
plt.xlabel("Cantidad de registros")
plt.ylabel("País")
plt.show()
```
